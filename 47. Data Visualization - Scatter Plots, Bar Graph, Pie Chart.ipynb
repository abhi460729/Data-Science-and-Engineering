### Scatter Plots

```python
import numpy as np
import matplotlib.pyplot as plt
prices = np.array([1,2,3,4])**3 #The indices becomes x-axis points
print(prices)
x = np.arange(10)
y1 = x**2
y2 = 2*x+3
```

```python
plt.plot(prices) #This becomes y-axis #original format plt.plot(x,y)
plt.show()
```

### Scatter Plots

```python
plt.scatter(x,y1)
plt.scatter(x,y2)
plt.show()
```

```python
#Scatter Plot
plt.scatter(x,y1,color='red',label="Apple",marker='o')
plt.scatter(x,y2,color='green',label="Kiwi",linestyle='dashed',marker='*')#jupyter notebook has the ability to show the graph but ni will not be shown in case of general python file without show().
plt.xlabel("Time")
plt.ylabel("Price")
plt.title("Prices of Fruits over Time")
plt.legend()
plt.show()
```

### Bar Graphs

```python
plt.bar([0,1,2],[10,20,15]) #(X,Y) #Current Year
plt.bar([0,1,2],[20,10,12]) # Next Year
plt.show()
```

```python
#I want these bars to be non-overlapping
x_coordinates = np.array([0,1,2])
plt.bar([0,1,2],[10,20,15]) #(X,Y) #Current Year
plt.bar(x_coordinates+0.5,[20,10,12]) # Next Year
plt.show()
```

```python
#Now there is an overlapping between two different bars, let's remove it.
x_coordinates = np.array([0,1,2])
plt.bar([0,1,2],[10,20,15], width = 0.5) #(X,Y) #Current Year
plt.bar(x_coordinates+0.5,[20,10,12], width=0.5) # Next Year
plt.show()
```

```python
#Now we need spacing between the bars
plt.style.use("seaborn")
x_coordinates = np.array([0,1,2])*2
plt.bar(x_coordinates-0.25,[10,20,15], width = 0.5,label="Current Year", tick_label=["Gold","Platinum","Silver"]) #(X,Y) #Current Year
plt.bar(x_coordinates+0.25,[20,10,12], width=0.5,label="Next Year") # Next Year
plt.title("Metal Price Comparison")
plt.xlabel("Metal")
plt.ylabel("Price")
plt.legend()
plt.show()
```

```python
#Now we need spacing between the bars
plt.style.use("dark_background")
x_coordinates = np.array([0,1,2])*2
plt.bar(x_coordinates-0.25,[10,20,15], width = 0.5,label="Current Year", tick_label=["Gold","Platinum","Silver"],color="green") #(X,Y) #Current Year
plt.bar(x_coordinates+0.25,[20,10,12], width=0.5,label="Next Year",color="orange") # Next Year
plt.title("Metal Price Comparison")
plt.xlabel("Metal")
plt.ylabel("Price")
plt.ylim(0,40)
plt.xlim(-2,5)
plt.legend()
plt.show()
```

### Pie Chart

```python
subjects = "Maths","Chem","Physics","English"
weightage = [20, 10, 15, 5]
plt.pie(weightage, labels = subjects)
plt.show()
```

```python
subjects = "Maths","Chem","Physics","English"
weightage = [20, 10, 15, 5]
plt.pie(weightage, labels = subjects, explode=(0,0,0.1,0))
plt.show()
```

```python
subjects = "Maths","Chem","Physics","English"
weightage = [20, 10, 15, 5]
plt.pie(weightage, labels = subjects, explode=(0,0,0.1,0),autopct = '%1.1f%%')
plt.show()
```

```python
plt.style.use("seaborn")
subjects = "Maths","Chem","Physics","English"
weightage = [20, 10, 15, 5]
plt.pie(weightage, labels = subjects, explode=(0,0,0.1,0),autopct = '%1.1f%%',shadow="True")
plt.title("Sbjects")
plt.show()
```
