```python
import cv2
import matplotlib.pyplot as plt
```

```python
img = cv2.imread(r'C:\Users\LENOVO\OneDrive\Desktop\Pictures\bulbasaur.jpg')
rgb_image = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
# BGR to RGB
# Original image is in RGB
# openCV stores image as BGR but we can convert it as above to RGB
```

```python
print(img.shape)
```

```python
#crop_image =rgb_image[rows,columns,]
crop_image = rgb_image[200:800,600:1700,:]
plt.imshow(rgb_image)
plt.axis("off") #to turn off the axis
plt.show()
```

```python
plt.imshow(crop_image)
plt.axis("off") #to turn off the axis
plt.show()
```
