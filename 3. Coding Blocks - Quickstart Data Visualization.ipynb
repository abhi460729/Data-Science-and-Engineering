```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.array([1,2,3,4,5])
y = x**2+3
print(x,y)
```

```python
plt.style.available
# Gives all the styles available for plotting
```

```python
plt.style.use('seaborn')
# To add a specific style
```

```python
plt.plot(x,x**2+x,label ="Stock Price") # Using the line function plot to plot the data points between x and y
plt.plot(x,x**2,label ="Stock Price") # Using the second line function plot to plot the data points between x and y
plt.title("My Line Plot") # To add a title
plt.xlabel("Timestamp") # To add x-label
plt.ylabel("Price") # To add y-label
plt.legend() # To show what the line represents i.e, it represents the Stock Price
plt.show() # To show the plot
```

```python
plt.scatter(x,x**2+x,label ="Stock Price") # Using the line function plot to plot the data points between x and y
plt.scatter(x,x**2,label ="Stock Price") # Using the second line function plot to plot the data points between x and y
plt.title("My Line Plot") # To add a title
plt.xlabel("Timestamp") # To add x-label
plt.ylabel("Price") # To add y-label
plt.legend() # To show what the line represents i.e, it represents the Stock Pric
plt.show() # To show the plot
```
