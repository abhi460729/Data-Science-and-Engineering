```python
import bs4
import requests
```

```python
url = "https://www.passiton.com/inspirational-quotes?page=2"
```

```python
response = requests.get(url)
```

```python
response.content
```

```python
# HTML parser
soup = bs4.BeautifulSoup(response.content)
```

```python
section_element = soup.findAll('section')
```

```python
div_element = section_element[1].div
```

```python
inner_div_element = div_element.findAll('div')
```

```python
div = inner_div_element[4].findAll('div')
```

```python
div
```

```python
for i, div in enumerate(div):
    with open('inspiration{}.jpg'.format(i),'wb') as file:
        img_url = div.a.img.attrs['src']
        response = requests.get(img_url)
        file.write(response.content)
```
