```python
import seaborn as sns
import numpy as np
```

### DISTRIBUTION PLOT

```python
#Till now we have done univariant and bivariant analysis.
```

```python
#It looks like a histogram plot but it provides more information than a histogram plot.
sns.distplot(tips_data["tip"], bins =20) # kernel density estimation, it is used when you have a sample from an unknown distribution and you can kind of approximate the probability density function of that unknown distribution.
```

### Heatmap

```python
# It is used when we want to see the distribution within the data frame.
```

```python
flights = sns.load_dataset('flights')
```

```python
flights
```

```python
# Show me the distribution of the passengers who actually flew in an airplane according to the months for different years.
```

```python
flights_pivot = flights.pivot_table(index="month", columns="year", values="passengers")
```

```python
# Give me the range of months and year as well for which the maximum number of passengers flow accross the airplane.
flights_pivot
```

```python
# For different months get the maximum value and then select the top three values or top 4 months for which maximum passengers were present.
```

```python
# Give me the visualization of the distribution of the passengers across different years and across different months.
```

```python
sns.heatmap(flights_pivot)
```

```python
# We can see a gradient going from left to right. As and when the advancement happened.Jun, July and Aug are actually the months that saw a lot of passengers.
```

```python
# There is a matrix and you want to do the visualization of the data in the matrix then heatmap is used.
```

```python
# We can also do this for correlation matrix using a function called df.corr()
```

```python
# It is needed when we need to look at the distribution of the data in the matrix.
```

```python
tips_data.corr()
```

```python
#It is going to give me the correlation between all the non categorical variables or the numerical variables.
```

```python
sns.heatmap(tips_data.corr())
```
